```python
import re

loc = soup.find_all('div', class_='companyCardWrapper__interLinkingWrapper')

for i in loc:
    text = i.text
    
    # Extract the industry
    industry_match = re.search(r'([A-Za-z\s&]+) \|', text)
    industry = industry_match.group(1) if industry_match else ""
    
    # Extract the number of employees
    employees_match = re.search(r'(\d+\s?[A-Za-z]+\sEmployees)', text)
    employees = employees_match.group(1) if employees_match else ""
    
    # Extract the location
    location_match = re.search(r'\| (\d+\s?[A-Za-z\s/]+ \+\d+\s?more)', text)
    location = location_match.group(1) if location_match else ""
    
    # Print the extracted values
    print("Industry:", industry)
    print("Number of Employees:", employees)
    print("Location:", location)
    print()
```

```python
salary_list = []

for i in rew:
    text = i.text
    
     # Check if the text contains the word "Salaries"
    if "Salaries" in text:
        # Extract the numeric part of the text (e.g., "733.7k Salaries" -> "733.7k")
        salary_match = re.search(r'([\d.]+k) Salaries', text)
        if salary_match:
            salary_value = salary_match.group(1)
            salary_list.append(salary_value)

# Print the list of salary values
print(salary_list)
    
```

```python
# Initialize an empty list to store job values
job_list = []

for i in rew:
    text = i.text
    
    # Check if the text contains the word "Jobs"
    if "Jobs" in text:
        # Extract the numeric part of the text (e.g., "45 Jobs" -> "45")
        job_match = re.search(r'(\d+) Jobs', text)
        if job_match:
            job_value = job_match.group(1)
            job_list.append(job_value)

# Print the list of job values
print(job_list)
```
