```python
#Prepare data:We will assume we have a single input feature y=x/2+sin(x)+e
#here e = Gaussian noise -N(0,1)

no_of_samples = 100 # sets the number of data samples you want to generate.

x = np.linspace(-np.pi, np.pi,no_of_samples) #generating an array of no_of_samples evenly spaced values between -π and π using the linspace function from the NumPy library.
y = 0.5*x+np.sin(x)+np.random.random(x.shape) #generates the corresponding output values (y) for each input value (x)

#0.5*x: This is a linear term that grows with x.
#np.sin(x): This is a sine wave term that introduces a periodic component.
#np.random.random(x.shape): This adds some random noise to the data, making it more realistic and capturing the variability that real-world data often exhibits.

#view the created data
for i in range(no_of_samples):
    print(f"x:{x[i]} y:{y[i]}")

# Display the top 5 and bottom 5 values of the created data
top_values = list(zip(x[:5], y[:5]))
bottom_values = list(zip(x[-5:], y[-5:]))

print("Top 5 values:")
for x_val, y_val in top_values:
    print(f"x: {x_val:.4f}, y: {y_val:.4f}")    
    
    
#visualize the data...
plt.scatter(x,y,color="darkblue") #plot yVSx in dots
plt.xlabel('x-input feature')
plt.ylabel('y-target values')
plt.title('FIG 1 : Data for Linear Regression')
plt.show()
```

```python
#Split the data into training,validation and test sets
random_indices = np.random.permutation(no_of_samples)
#Training Set
x_train = x[random_indices[:70]]
y_train = y[random_indices[:70]]
#Validation set
x_val = x[random_indices[70:85]]
y_val = y[random_indices[70:85]]
#Test set
x_test = x[random_indices[85:]]
y_test = y[random_indices[85:]]

# Display a sample of the training data
for i in range(len(x_train)):
    print(f"x_train: {x_train[i]:.4f}, y_train: {y_train[i]:.4f}")

plt.scatter(x_train, y_train, color="darkblue")
plt.xlabel('x-input feature')
plt.ylabel('y-target values')
plt.title('Training Data')
plt.show()
```

```python
# Create a LinearRegression model
#Linear regression learns to fit a hyperplane to our data in the feature space. For one dimensional data, the hyperplane reduces to a straight line. We will fit a line to our data using sklearn.linear_model.LinearRegression

model = linear_model.LinearRegression() #Create a least squared error linear regression object

#sklearn takes the inputs as matrices. Hence we reshpae the arrays into column matrices
x_train_for_line_fitting = np.matrix(x_train.reshape(len(x_train),1))
y_train_for_line_fitting = np.matrix(y_train.reshape(len(y_train),1))


```

```python
# Fit the line to the training data
model.fit(x_train.reshape(-1, 1), y_train)

# Plot the line
plt.scatter(x_train, y_train, color='black', label='Training Data')
plt.plot(x, model.predict(x.reshape(-1, 1)), color='blue', linewidth=2, label='Fitted Line')
plt.xlabel('x-input feature')
plt.ylabel('y-target values')
plt.title('Fig 2: Line fit to training data')
plt.legend()
plt.show()
```
